```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['Person']= titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

Q.2 what deck the passngers were on and how it relates to the pclass?

```python
titanic_df.head()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)

cabin_df.columns= ['Cabin']

sns.catplot('Cabin',data=cabin_df,palette='winter_d',kind='count')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T' ]

sns.catplot('Cabin',data=cabin_df,palette='summer',kind='count')
```

Q.4. who was alone and who was with family?

```python
titanic_df.head()
```

```python
titanic_df['Alone'] = titanic_df.Parch + titanic_df.SibSp
```

```python
titanic_df['Alone'].head()
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'

titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'alone'
```

```python
titanic_df[0:10]
```

```python
sns.catplot('Alone',data=titanic_df,palette='Blues',kind='count')
```
